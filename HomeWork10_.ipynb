```python
from google.colab import files

uploaded = files.upload()

# Припустимо, що файл завантажено як 'movies.mat'
import scipy.io

data = scipy.io.loadmat('movies.mat')
print("Дані успішно завантажені.")
```

```python
# Завантаження даних
ratings = data['Y']
R = data['R']
```

```python
import numpy as np
from scipy.optimize import minimize

# Визначення функцій
def cost_function(params, Y, R, num_users, num_movies, num_features, lambda_):
    X = params[:num_movies*num_features].reshape(num_movies, num_features)
    Theta = params[num_movies*num_features:].reshape(num_users, num_features)

    error = np.dot(X, Theta.T) - Y
    error = error * R
    cost = (1/2) * np.sum(np.square(error))

    cost += (lambda_/2) * (np.sum(np.square(Theta)) + np.sum(np.square(X)))

    return cost

def gradient(params, Y, R, num_users, num_movies, num_features, lambda_):
    X = params[:num_movies*num_features].reshape(num_movies, num_features)
    Theta = params[num_movies*num_features:].reshape(num_users, num_features)

    error = np.dot(X, Theta.T) - Y
    error = error * R

    X_grad = np.dot(error, Theta) + lambda_ * X
    Theta_grad = np.dot(error.T, X) + lambda_ * Theta

    grad = np.concatenate([X_grad.ravel(), Theta_grad.ravel()])

    return grad

# Ініціалізація параметрів
num_movies, num_users = ratings.shape
num_features = 10

X = np.random.randn(num_movies, num_features)
Theta = np.random.randn(num_users, num_features)

initial_params = np.concatenate([X.ravel(), Theta.ravel()])

lambda_ = 10

# Оптимізація параметрів
result = minimize(fun=cost_function, x0=initial_params, args=(ratings, R, num_users, num_movies, num_features, lambda_),
                  method='CG', jac=gradient, options={'maxiter': 100})

learned_params = result.x
X = learned_params[:num_movies*num_features].reshape(num_movies, num_features)
Theta = learned_params[num_movies*num_features:].reshape(num_users, num_features)
```
